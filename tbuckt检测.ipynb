```python
from htrace import Parse
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

%config InlineBackend.figure_formats = ['svg']
sns.set_style("whitegrid")
```

```python
always = Parse('/Users/yangs/Desktop/ka.out', True)
tbuck = Parse('/Users/yangs/Desktop/kt.out', True)
```

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(2, 1, figsize = (10, 8))
def timedata(a):
    at = a.df
    at = at[at['name']=='DFSOutputStream#writeChunk']
    at['begin'] = at['begin'] - at['begin'].min()
    return at
charts = [always,tbuck]
for i in range(2):
    sns.distplot(timedata(charts[i])['begin'], bins=100, ax=axes[i], kde=False)
```
