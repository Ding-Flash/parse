## 特定函数(次数不多)性能影响

```python
def clean_src_csv(path):
    df = pd.read_csv(path+'/1.csv')
    df.index = df['Unnamed: 0']
    df.drop(['Unnamed: 0'],axis=1, inplace=True)
    df = df.T
    df =df.drop(['25%','50%','min','75%'], axis=1)
    df.columns = list(map(lambda x: path.split('/')[2]+'-'+x,['count','mean','std','max']))
    return df
```

```python
def bench_func_display(bench):
    res = []
    for sampler in samplers:
        res.append(clean_src_csv(bench+sampler))
    c = pd.concat(res, axis=1, sort=True)
#     c.columns = zh_samplers
    c.insert(0, 'time', c['a-count']*c['a-mean'])
    return c[(c['a-std']>c['a-mean']+1)&(c['a-count']>100)&(c['l-std']>c['l-mean']+1)&(c['t-std']>c['t-mean']+1)]
```

```python
for name, bench in data_path.items():
    print(bench)
    bench_func_display(bench)
```
