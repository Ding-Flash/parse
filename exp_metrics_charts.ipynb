## 不同bench的采样器采样指标(图)

```python
def draw_chart(test):
    fig, axes = plt.subplots(2, 3, figsize = (12, 10))
    sns.barplot(x=test.index, y=test.node, ax=axes[0,0])
    sns.barplot(x=test.index, y=test.tree, ax=axes[0,1])
    sns.barplot(x=test.index, y=test.en, ax=axes[0,2])
    sns.barplot(x=test.index, y=test.dist, ax=axes[1,0])
    sns.barplot(x=test.index, y=test.time, ax=axes[1,1])
    test['size'] = [0]+test['size'].tolist()[1:]
    i = ['']+test.index.tolist()[1:]
    sns.barplot(x=i, y=test['size'], ax=axes[1,2])
    plt.subplots_adjust(wspace=0.3, hspace=0.15)
```

### dfsioe_r

```python
dfsioe_r = merge_info('data/dfsioe_r/', samplers)
draw_chart(dfsioe_r)
```

### dfsioe_w

```python
dfsioe_w = merge_info('data/dfsioe_w/', samplers)
draw_chart(dfsioe_w)
```

### kmeans
