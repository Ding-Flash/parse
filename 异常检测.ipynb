```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from htrace import Parse
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

%config InlineBackend.figure_formats = ['svg']
sns.set_style("whitegrid")
```

```python
data_path={
    'dfsioe_r': 'ndata/dfsioe_r/',
    'dfsioe_w': 'ndata/dfsioe_w/',
    'kmeans': 'ndata/kmeans/',
    'pagerank': 'ndata/pagerank/',
    'terasort': 'ndata/terasort/',
    'wordcount': 'ndata/wordcount/'
}
```

```python
def clean_src_csv(path):
    df = pd.read_csv(path)
    df.index = df['Unnamed: 0']
    df.drop(['Unnamed: 0'],axis=1, inplace=True)
    df = df.T
    df =df.drop(['min','25%'], axis=1)
    df.columns = list(map(lambda x: path.split('/')[2]+'-'+x,['count','mean','std','50%','75%','max']))
    return df
```

### wordcount

```python
bench = 'data/wordcount/'
ans = []
for i in range(1):
    c = clean_src_csv(bench+'l/'+str(i+1)+'.csv')
    c = c[c['l-count']>10]
    tar = pd.read_csv(bench+'l/'+str(i+1)+'s.csv')
    err = []
    x = []
    for i in c.iterrows():
        d = tar[tar['name']==i[0]]
        res = 0
        for j in d.iterrows():
            if j[1]['time'] > 1.5*i[1]['l-50%']:
                res += 1
        err.append((res/i[1]['l-count'])*100)
        x.append(res)
    c['err%'] = err
    c['res'] = x
    ans.append(c[
        (c['err%']<30)
        &(c['err%']>0)
        &(c['l-max']>1)
    ].sort_values('err%'))
c = pd.concat(ans, axis=1)
c.dropna(axis=0,how='any')
c.to_csv('/Users/yangs/Desktop/wordcount.csv')
```
